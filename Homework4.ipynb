```python
# Unique Email Addresses
# https://leetcode.com/problems/unique-email-addresses/description/ 

def numUniqueEmails(emails):
    unique_emails=set()
    for email in emails:
        name, domain= email.split('@')
        name=name.replace('.','')
        if '+' in name:
            name=name.split('+')[0]
        unique_emails.add(name+'@'+domain)
    return len(unique_emails)

emails = ["test.email+alex@leetcode.com","test.e.mail+bob.cathy@leetcode.com","testemail+david@lee.tcode.com"]

numUniqueEmails(emails)
```

```python
# Find and Replace Pattern
# https://leetcode.com/problems/find-and-replace-pattern/description/ 

def findAndReplacePattern(words, pattern):
    word_char_index=[]
    mapped_words=[]
    pattern_char_index=[pattern.index(ch) for ch in pattern ]
    for word in words:
        word_char_index=[word.index(ch) for ch in word] 
        if word_char_index==  pattern_char_index:
            mapped_words.append(word)
    return mapped_words

words = ["deq","abc","mee","aqq","dkd","ccc"]
pattern = "frd"

findAndReplacePattern(words, pattern)
```

```python
# Beautiful Binary String
# https://www.hackerrank.com/challenges/beautiful-binary-string/problem

def beautifulBinaryString(b):
    substr='010'
    n=0
    b_list=list(b)
    while b.find(substr)>=0:
        b_list[b.find(substr)+2]='1'
        b=''.join(b_list)
        n+=1
    return n


b = '0101010'
beautifulBinaryString(b)       
```

```python
# Jewels and Stones
#https://leetcode.com/problems/jewels-and-stones/description/ 

def numJewelsInStones(jewels, stones):
        count=0
        for j in jewels:
            for i in stones:
                if j == i:
                    count+=1
        return count
    
    
jewels = "aA"
stones = "aAAbbbb"

numJewelsInStones(jewels, stones)    
```

```python
# String power

def string_power(s,k):
    if k==0:
        x_string=''
    elif k>0:
        x_string=s*k
    elif len(s)%k!=0:
        x_string='undefined'
    else:
        rep=int(len(s)/abs(k))
        for i in range(0,len(s)-rep+1,rep):
            
            if (s[:rep]!=s[i:i+rep]):
                x_string='undefined'
            else:
                x_string=(s[i:i+rep])
            
    return x_string    

string_power('asasas', -3)        
```
