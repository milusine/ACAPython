```python
#Digit Product

def digit_Product(n):
    result=1
    while n>0:
        digit=n%10
        if digit!=0:
            result*=digit      
        n//=10
    return result
n=int(input())
digit_product(n)    
```

```python
# Largest Power of 3

def largest_power_3(n):
    result=1
    while result<=n:
        result*=3
    return int(result/3)
    
n=int(input())
largest_power_3(n)    
```

```python
# Triangle

def Triangle(a,b,c):
    min_side=min(a,b,c)
    max_side=max(a,b,c)
    mid_side=a+b+c-min_side-max_side
    if max_side>=min_side+mid_side:
        return 0
    elif max_side**2==min_side**2+mid_side**2:
        return 1
    elif max_side**2<min_side**2+mid_side**2:
        return 2
    else:
        return 3

a=int(input())
b=int(input())
c=int(input())
if Triangle(a,b,c)==1:
    print('Right Triangle')
elif Triangle(a,b,c)==2:
        print('Acute Triangle')
elif Triangle(a,b,c)==0:
    print('No Triangle') 
else:
    print('Obtuse Triangle')       
```

```python
# The Root of the Number

def root(n):
    while n>10:
        result=0
        while n>0:
            result+=n%10
            n//=10
        n=result
        print(result)

    
n=int(input())
root(n)        
```

```python
#Number Of Divisors

def n_divisors(num):
    n=0
    for i in range(1,int(num/2)+1):
        if num%i==0:
            n+=1
    return n+1
n=int(input())
print(n_divisors(n))
            
```
