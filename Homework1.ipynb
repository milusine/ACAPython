```python
#Arithmetic Progression
a1=int(input())
a2=int(input())
n=int(input())
an=a1+(a2-a1)*(n-1)
print(an)
```

```python
# Century from year
year=int(input())
century=int((year-1)/100+1)
print(century)

```

```python
# Two man
first_shots=int(input())
second_shots=int(input())
all_cans=first_shots+second_shots-1
first_missed=all_cans-first_shots
second_missed=all_cans-second_shots
print(first_missed, second_missed)
```

```python
# Knight's Possible Moves
px=int(input())
py=int(input())
print(px+2, py-1)
print(px+2, py+1)
print(px-2, py-1)
print(px-2, py+1)
print(px+1, py+2)
print(px-1, py+2)
print(px+1, py-2)
print(px-1, py-2)
```
