```python
#Goldbach's Conjecture

def n_divisors(num):
    n=0
    for i in range(1, int(num/2)+1):
        if num%i==0:
            n+=1
    return n+1

def sum_of_primes(n):
    for a in range(1,n,2):
        if (a==1 or n_divisors(a)==2) and n_divisors(n-a)==2:
            return a, n-a

n=int(input())        
print(sum_of_primes(n))
         
```

```python
#Polindrome numbers

def reverse(n):
    reversed_n=n%10
    n//=10
    while n>0:
        reversed_n*=10
        reversed_n+=n%10
        n//=10
    return reversed_n

def polindrome(num1, num2):
    list_of_polindromes=[]
    for i in range(num1,num2):
        if i==reverse(i):
            list_of_polindromes.append(i)
    return list_of_polindromes

n1=int(input())
n2=int(input())
print(polindrome(n1,n2))        
```

```python
#Suffix Sums

def suffix_sums(a):
    sums_after=[]
    sums_before=0
    sums_after.append(sum(a))
    for i in range(1,len(a)):
        sums_before+=a[i]
        sums_after.append(sums_after[i-1]-a[i-1])
    return sums_after

a=[1,2,3,4,5]
print(suffix_sums(a))
```

```python
#Cyclic shift4

def cyclic_shift(n,k, seq):
    for _ in range(k):
        seq=[seq.pop(n-1)]+seq
    return seq


n=6
k=2
seq=[1, 2, 3, 4, 5, 6]
print(cyclic_shift(n,k, seq))
    


    
    
```

```python
#Tree

def tree(n):
    t=[]
    j=1    
    for i in range(round(n/2+1)):

        t.append(''.join([' ' for _ in range(round(n/2)-i)]))
        t[i]+=''.join(['*' for _ in range(i+j) if i+j<=n])
        j+=1

    for line in t:
        print(line)

n=int(input('Enter an odd number'))
tree(n)
```
